```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
dat = pd.read_csv('/home/maria/CrimeStats/data/avalik_1.csv', sep='\t')

# Filter for crimes in Tallinn
tallinn = dat[dat['ValdLinnNimetus'] == 'Tallinn']

# Combine date and time columns into a single datetime column
# Parse the 'ToimKpv' as a date
tallinn['ToimKpv'] = pd.to_datetime(tallinn['ToimKpv'], errors='coerce')

# For 'ToimKell', we'll fill NaN values with '00:00' and then parse it as time
tallinn['ToimKell'] = tallinn['ToimKell'].fillna('00:00')

# Convert 'ToimKell' to a proper timedelta (hours and minutes)
tallinn['ToimKell'] = pd.to_timedelta(tallinn['ToimKell'] + ':00')  # Append ':00' to get HH:MM:SS format

# Create a new column by adding the date and time together
tallinn['ToimDatetime'] = tallinn['ToimKpv'] + tallinn['ToimKell']

# Drop rows where the datetime could not be parsed
tallinn = tallinn.dropna(subset=['ToimDatetime'])

# Set the index to the new datetime column
tallinn.set_index('ToimDatetime', inplace=True)

# Bin the data by day and count the number of crimes per day
crimes_per_day = tallinn.resample('D').size()

# Plot the time series of crimes per day
plt.figure(figsize=(10, 6))
plt.plot(crimes_per_day.index, crimes_per_day.values, label='Crimes per Day', color='blue')
plt.title('Crimes per Day in Tallinn')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from scipy import stats

# Function to fit an exponential mixture model using a Gaussian Mixture Model on log-transformed data
def fit_exponential_mixture(inter_arrival_times, n_components=2):
    # Take the log of inter-arrival times since exponentials become linear in log-space
    log_times = np.log(inter_arrival_times).reshape(-1, 1)
    
    # Fit a Gaussian Mixture Model on the log-transformed data
    gmm = GaussianMixture(n_components=n_components)
    gmm.fit(log_times)
    
    # Extract the means and variances of the Gaussians (which correspond to rate parameters in log-space)
    exp_means = np.exp(gmm.means_.flatten())
    exp_weights = gmm.weights_
    
    return exp_means, exp_weights, gmm

# Assuming `tallinn` data is already processed and indexed by 'ToimDatetime'

# Sort the DataFrame by datetime to ensure proper calculation of time differences
tallinn = tallinn.sort_index()

# Calculate time differences (inter-arrival times) in days between consecutive crimes
inter_arrival_times_days = tallinn.index.to_series().diff().dt.total_seconds().dropna() / (60 * 60 * 24)  # Convert seconds to days

# Fit the mixture model to the inter-arrival times
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Function to fit a mixture of two exponential distributions using EM
def fit_exponential_mixture_EM(inter_arrival_times, n_components=2, max_iter=100, tol=1e-6):
    # Initialize parameters
    np.random.seed(42)
    n = len(inter_arrival_times)
    lambdas = np.random.uniform(0.5, 2.0, n_components)  # Initialize random lambda (rate) values
    weights = np.ones(n_components) / n_components  # Initialize equal weights
    
    log_likelihoods = []
    
    for iteration in range(max_iter):
        # E-step: Compute responsibilities (posterior probabilities)
        responsibilities = np.zeros((n, n_components))
        for j in range(n_components):
            responsibilities[:, j] = weights[j] * stats.expon.pdf(inter_arrival_times, scale=1/lambdas[j])
        responsibilities /= responsibilities.sum(axis=1, keepdims=True)
        
        # M-step: Update weights and lambda (rate) values
        weights = responsibilities.mean(axis=0)
        lambdas = np.array([
            (responsibilities[:, j] * inter_arrival_times).sum() / responsibilities[:, j].sum()
            for j in range(n_components)
        ])
        lambdas = 1 / lambdas  # Convert back to lambda (rate)
        
        # Compute log likelihood
        log_likelihood = np.sum(np.log(responsibilities.sum(axis=1)))
        log_likelihoods.append(log_likelihood)
        
        # Check for convergence
        if iteration > 0 and np.abs(log_likelihoods[-1] - log_likelihoods[-2]) < tol:
            break
    
    return lambdas, weights, log_likelihoods

# Assuming `tallinn` data is already processed and indexed by 'ToimDatetime'

# Sort the DataFrame by datetime to ensure proper calculation of time differences
tallinn = tallinn.sort_index()

# Calculate time differences (inter-arrival times) in days between consecutive crimes
inter_arrival_times_days = tallinn.index.to_series().diff().dt.total_seconds().dropna() / (60 * 60 * 24)  # Convert seconds to days

# Fit the mixture of two exponential distributions using EM
exp_lambdas, exp_weights, log_likelihoods = fit_exponential_mixture_EM(inter_arrival_times_days.values, n_components=2)

# Plot the histogram of the inter-arrival times with KDE
plt.figure(figsize=(10, 6))
sns.histplot(inter_arrival_times_days, bins=50, kde=True, color='blue')
plt.title('Histogram of Inter-Arrival Times Between Crimes (Days)')
plt.xlabel('Time Between Crimes (days)')
plt.ylabel('Frequency')
plt.grid(True)

# Generate some data from the learned mixture model for visualization
x_vals = np.linspace(0, inter_arrival_times_days.max(), 1000)
mixture_pdf = np.sum([w * stats.expon.pdf(x_vals, scale=1/lamb) for w, lamb in zip(exp_weights, exp_lambdas)], axis=0)
plt.plot(x_vals, mixture_pdf * len(inter_arrival_times_days) * np.diff(x_vals)[0], color='red', label='Mixture Model PDF')
plt.legend()

plt.tight_layout()
plt.show()

# Q-Q plot for the mixture model vs the data
# Sample synthetic inter-arrival times from the mixture model
n_samples = len(inter_arrival_times_days)
sampled_intervals = np.concatenate([
    np.random.exponential(scale=1/lamb, size=int(n_samples * w))
    for lamb, w in zip(exp_lambdas, exp_weights)
])

# Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(inter_arrival_times_days, dist="expon", plot=plt)
plt.title('Exponential Q-Q Plot of Mixture Model vs Data')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print learned parameters of the mixture model
print("Learned Exponential Mixture Model Parameters:")
for i, (lamb, weight) in enumerate(zip(exp_lambdas, exp_weights)):
    print(f"Component {i + 1}: Lambda = {lamb:.4f} (Rate), Weight = {weight:.4f}")
exp_means, exp_weights, gmm = fit_exponential_mixture(inter_arrival_times_days.values)

# Plot the histogram of the inter-arrival times with KDE
plt.figure(figsize=(10, 6))
sns.histplot(inter_arrival_times_days, bins=50, kde=True, color='blue')
plt.title('Histogram of Inter-Arrival Times Between Crimes (Days)')
plt.xlabel('Time Between Crimes (days)')
plt.ylabel('Frequency')
plt.grid(True)

# Generate some data from the learned mixture model for visualization
x_vals = np.linspace(0, inter_arrival_times_days.max(), 1000)
mixture_pdf = np.sum([w * stats.expon.pdf(x_vals, scale=mean) for w, mean in zip(exp_weights, exp_means)], axis=0)
plt.plot(x_vals, mixture_pdf * len(inter_arrival_times_days) * np.diff(x_vals)[0], color='red', label='Mixture Model PDF')
plt.legend()

plt.tight_layout()
plt.show()

# Q-Q plot for the mixture model vs the data
# Sample synthetic inter-arrival times from the mixture model
n_samples = len(inter_arrival_times_days)
sampled_intervals = np.concatenate([
    np.random.exponential(scale=mean, size=int(n_samples * w))
    for mean, w in zip(exp_means, exp_weights)
])

# Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(inter_arrival_times_days, dist="expon", sparams=(1/np.mean(inter_arrival_times_days)), plot=plt)
plt.title('Exponential Q-Q Plot of Mixture Model vs Data')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print learned parameters of the mixture model
print("Learned Exponential Mixture Model Parameters:")
for i, (mean, weight) in enumerate(zip(exp_means, exp_weights)):
    print(f"Component {i + 1}: Mean = {mean:.2f} days, Weight = {weight:.2f}")
```
